```python
# Your code here
unique_names = data['gender'].unique()
unique_names
```

```python

def clean(x):
    if x in ['M','Male']:
        return 'M'
    elif x in ['F','Femal', 'female']:
        return 'F'
    else:
        return np.nan
    

data['gender'] = data['gender'].apply(clean)
data
```

```python
def clean(x):
    if x in ['WA','Washington']:
        return 'Washington'
    elif x in ['AZ', 'Arizona']:
        return 'Arizona'
    elif x in ['Nevada']:
        return 'Nevada'
    elif x in ['California', 'Cali']:
        return 'California'
    elif x in ['Oregon']:
        return 'Oregon'
    else:
        return np.nan
    

data['state'] = list(map(clean, data['state']))
data
```

```python
data['vehicle_class	']=data['education'].replace({'Bachelors':'Bachelor'})
data['education'].unique()
```
